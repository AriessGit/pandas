```python
# 1. Sorting ðŸ”¢

# Q1.1 Sort the DataFrame by mpg in ascending order
display('DataFrame by mpg in ascending order', df.sort_values(by = 'mpg',ascending = True).head(5))

# Q1.2 Sort the DataFrame by weight in descending order
display('DataFrame by weight in descending order', df.sort_values('weight', ascending = False).head(5))

# Q1.3 Sort by model_year first, then by mpg within each year
display('Sort by model_year first, then by mpg within each year',df.sort_values(['model_year', 'mpg']))

```

```python
# Value Counts ðŸ“Š

df = pd.read_csv("mpg.csv")

# Q2.1 Find how many cars of each cylinder type exist
display(df.value_counts('cylinders', ascending = True).to_frame())
print()
# for practice, I've written another commands: 
display(df['cylinders'].value_counts().to_frame()) # Ascending = False -- maximum  goe's at first
print()
display(df['cylinders'].value_counts().sort_index()) # sorted by index 
print()

# Q2.2 Count how many times each origin appears
display(df['origin'].value_counts().sort_index())
print()
# Q2.3 Which model_year has the most cars?
display(df['model_year'].value_counts().head(1))
```

```python
#GroupBy & Aggregations ðŸ§®
# Q3.1 Group by cylinders and calculate the average mpg
display(df.groupby(by = 'cylinders').mean('mpg'))

# Q3.2 Group by origin and find the maximum horsepower
display(df.groupby(by = 'origin').max('horspower'))

# Q3.3 Group by model_year and calculate both the mean and median of weight
display(df.groupby(by = 'model_year')['weight'].agg(['mean', 'median']))
```
